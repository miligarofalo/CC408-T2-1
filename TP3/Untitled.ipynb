Para cada a√±o, partan la base respondieron en una base de prueba
(test) y una de entrenamiento (train) utilizando el comando
train_test_split. La base de entrenamiento debe comprender el 70%
de los datos, y la semilla a utilizar (random state instance) debe ser 101.
Establezca a desocupado como su variable dependiente en la base de
entrenamiento (vector y). El resto de las variables ser√°n las variables
independientes (matriz X). Recuerden agregar la columna de unos (1).

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# BASE PARA CADA A√ëO

respondieron_2004 = respondieron[respondieron['anio'] == "2004"]
# Filtrar los datos para el a√±o 2024
respondieron_2024 = respondieron[respondieron['anio'] == "2024"]
```

```python
print(respondieron_2004.dtypes)
```

```python
def dividir_datos(data):
    # Filtrar para obtener solo las filas donde estado = 2 (desocupado)
    data_desocupados = data[data['estado'] == 2]

    # Definir la variable dependiente (y) y las independientes (X)
    y = data_desocupados['estado']  # Esto ser√° 2 para los desocupados
    X = data_desocupados.drop(columns=['estado'])  # Eliminar la columna dependiente de las independientes

    # Agregar una columna de unos para el intercepto
    X = X.assign(intercept=1)

    # Dividir los datos en conjunto de entrenamiento (70%) y prueba (30%)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

    return X_train, X_test, y_train, y_test

# Dividir los datos para 2004 y 2024
X_train_2004, X_test_2004, y_train_2004, y_test_2004 = dividir_datos(respondieron_2004)
X_train_2024, X_test_2024, y_train_2024, y_test_2024 = dividir_datos(respondieron_2024)
```

```python
X_train_2004 = pd.get_dummies(X_train_2004, drop_first=True)
X_train_2024 = pd.get_dummies(X_train_2024, drop_first=True)

X_test_2004 = pd.get_dummies(X_test_2004, drop_first=True)
X_test_2024 = pd.get_dummies(X_test_2024, drop_first=True)

# Asegurarse de que ambas matrices de caracter√≠sticas (X_train y X_test) tengan las mismas columnas
X_train_2004, X_test_2004 = X_train_2004.align(X_test_2004, join='left', axis=1, fill_value=0)
X_train_2024, X_test_2024 = X_train_2024.align(X_test_2024, join='left', axis=1, fill_value=0)

model_2004 = LogisticRegression()
model_2004.fit(X_train_2004, y_train_2004)

# Hacer predicciones en el conjunto de prueba para 2004
y_pred_2004 = model_2004.predict(X_test_2004)

# Evaluar el modelo para 2004
print("A√±o 2004 - Reporte de Clasificaci√≥n:")
print(classification_report(y_test_2004, y_pred_2004))
print("Matriz de Confusi√≥n:")
print(confusion_matrix(y_test_2004, y_pred_2004))

# Repite para el a√±o 2024
model_2024 = LogisticRegression()
model_2024.fit(X_train_2024, y_train_2024)

y_pred_2024 = model_2024.predict(X_test_2024)

print("A√±o 2024 - Reporte de Clasificaci√≥n:")
print(classification_report(y_test_2024, y_pred_2024))
print("Matriz de Confusi√≥n:")
print(confusion_matrix(y_test_2024, y_pred_2024))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve

# Definir funci√≥n para entrenar y evaluar el modelo
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]  # Para la curva ROC y AUC

    # Matriz de confusi√≥n y m√©tricas
    conf_matrix = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    
    # Curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f'{type(model).__name__} (AUC = {auc:.2f})')
    
    return conf_matrix, accuracy, auc

# Inicializar los modelos
models = {
    'Logistic Regression': LogisticRegression(),
    'Linear Discriminant Analysis': LinearDiscriminantAnalysis(),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=3),
    'Naive Bayes': GaussianNB()
}

# Evaluar los modelos para 2004 y 2024
results = {'2004': {}, '2024': {}}

for name, model in models.items():
    print(f"\nEvaluaci√≥n para el modelo {name} en 2004:")
    conf_matrix, accuracy, auc = evaluate_model(model, X_train_2004, X_test_2004, y_train_2004, y_test_2004)
    results['2004'][name] = {'conf_matrix': conf_matrix, 'accuracy': accuracy, 'auc': auc}
    print(f"Matriz de confusi√≥n:\n{conf_matrix}\nAccuracy: {accuracy:.2f}, AUC: {auc:.2f}")
    
    print(f"\nEvaluaci√≥n para el modelo {name} en 2024:")
    conf_matrix, accuracy, auc = evaluate_model(model, X_train_2024, X_test_2024, y_train_2024, y_test_2024)
    results['2024'][name] = {'conf_matrix': conf_matrix, 'accuracy': accuracy, 'auc': auc}
    print(f"Matriz de confusi√≥n:\n{conf_matrix}\nAccuracy: {accuracy:.2f}, AUC: {auc:.2f}")

# Mostrar la curva ROC para todos los modelos y a√±os
plt.figure(figsize=(10, 6))
plt.title("Curvas ROC para diferentes modelos (2004 y 2024)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.show()
```

```python
for year in results:
    print(f"\nResultados para el a√±o {year}:")
    for model in results[year]:
        print(f"Modelo: {model}")
        print(f"- Accuracy: {results[year][model]['accuracy']:.2f}")
        print(f"- AUC: {results[year][model]['auc']:.2f}")
        print(f"- Matriz de Confusi√≥n:\n{results[year][model]['conf_matrix']}\n")
```
